```
from transformers import pipeline
from IPython.display import display
import ipywidgets as widgets
```

```
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```
def smart_summarize(text):
    word_count = len(text.split())
    if word_count < 60:
        max_len, min_len = 40, 15
    elif word_count < 150:
        max_len, min_len = 80, 30
    elif word_count < 300:
        max_len, min_len = 120, 50
    else:
        max_len, min_len = 200, 80

    summary = summarizer(text, max_length=max_len, min_length=min_len, do_sample=False)
    return summary[0]['summary_text']
text_box = widgets.Textarea(
    value='',
    placeholder='Type or paste your paragraph here...',
    description='Input:',
    layout=widgets.Layout(width='100%', height='150px')
)

button = widgets.Button(description="Summarize", button_style='success')
output = widgets.Output()

def on_button_click(b):
    output.clear_output()
    with output:
        text = text_box.value
        if not text.strip():
            print(" Please enter some text.")
            return
        summary = smart_summarize(text)
        print(" Summary:\n", summary)

button.on_click(on_button_click)
display(text_box,button,output)
```
